```python
## Get the Geo data
nuts1 = Functions.fetch_nuts1_geojson()
nuts2 = Functions.fetch_nuts2_geojson()
nuts3 = Functions.fetch_nuts3_geojson()
```

```python
## Save the Geo data
json.dump(nuts1, 
          open("nuts1.json", 'w'))
json.dump(nuts2, 
          open("nuts2.json", 'w'))
json.dump(nuts3, 
          open("nuts3.json", 'w'))
```

```python
## Formatted version of Geo data NUTS1
countries = [country.alpha_2 for country in pycountry.countries]

country_features_nuts1 = {
    country_code : [feature for feature in nuts1["features"] if feature["properties"]["CNTR_CODE"] == country_code]
    for country_code in countries
}
country_features_nuts1_out = {i:j for i,j in country_features_nuts1.items() if len(j) != 0}

filtered_features_nuts1 = {
    country_code : Functions.exclude_islands_or_territories(cf, country_code)
    for country_code, cf in country_features_nuts1_out.items()
}

region_ids_nuts1 = {
    country_code : [feature["properties"]["NUTS_ID"] for feature in ff]
    for country_code, ff in filtered_features_nuts1.items()
}

region_names_nuts1 = {
    country_code : [feature["properties"]["NAME_LATN"] for feature in ff]
    for country_code, ff in filtered_features_nuts1.items()
}

## Save the formatted Geo data
json.dump(country_features_nuts1_out, 
          open("country_features_nuts1.json", 'w'))
json.dump(filtered_features_nuts1, 
          open("filtered_features_nuts1.json", 'w'))
json.dump(region_ids_nuts1, 
          open("region_ids_nuts1.json", 'w'))
json.dump(region_names_nuts1, 
          open("region_names_nuts1.json", 'w'))
```

```python
## Formatted version of Geo data NUTS2
country_features_nuts2 = {
    country_code : [feature for feature in nuts2["features"] if feature["properties"]["CNTR_CODE"] == country_code]
    for country_code in countries
}
country_features_nuts2_out = {i:j for i,j in country_features_nuts2.items() if len(j) != 0}

filtered_features_nuts2 = {
    country_code : Functions.exclude_islands_or_territories(cf, country_code)
    for country_code, cf in country_features_nuts2_out.items()
}

out_nuts2 = {
    i: {
        j: {
            k["properties"]["NUTS_ID"] : {
                "type" : "Feature",
                "properties" : k["properties"],
                "geometry" : k["geometry"]
            }
            for k in filtered_features_nuts2[i] if k["properties"]["NUTS_ID"][:3] == j[:3]
        }
        for j in region_ids_nuts1[i]
    }
    for i in filtered_features_nuts2.keys()
}

region_ids_nuts2 = {
    country_code : [feature["properties"]["NUTS_ID"] for feature in ff]
    for country_code, ff in filtered_features_nuts2.items()
}

out_nuts2_name = {
    i: {
        j: {
            k["properties"]["NAME_LATN"] : {
                "type" : "Feature",
                "properties" : k["properties"]
            }
            for k in filtered_features_nuts2[i] if k["properties"]["NUTS_ID"][:3] == j[:3]
        }
        for j in region_ids_nuts1[i]
    }
    for i in filtered_features_nuts2.keys()
}

region_names_nuts2 = {i : {
        k : list(out_nuts2_name[i][k].keys())
        for k in out_nuts2_name[i].keys()
    }
 for i in out_nuts2_name.keys()
 }

## Save the formatted Geo data
json.dump(country_features_nuts2, 
          open("country_features_nuts2.json", 'w'))
json.dump(out_nuts2, 
          open("filtered_features_nuts2.json", 'w'))
json.dump(region_ids_nuts2, 
          open("region_ids_nuts2.json", 'w'))
json.dump(region_names_nuts2, 
          open("region_names_nuts2.json", 'w'))
```

```python
## Formatted version of Geo data NUTS3
country_features_nuts3 = {
    country_code : [feature for feature in nuts3["features"] if feature["properties"]["CNTR_CODE"] == country_code]
    for country_code in countries
}
country_features_nuts3_out = {i:j for i,j in country_features_nuts3.items() if len(j) != 0}

filtered_features_nuts3 = {
    country_code : Functions.exclude_islands_or_territories(cf, country_code)
    for country_code, cf in country_features_nuts3_out.items()
}

out_nuts3 = {
    i : {
        j : {
            k : {
                l["properties"]["NUTS_ID"] : {
                    "type" : "Feature",
                    "properties" : l["properties"],
                    "geometry" : l["geometry"]
                }
                for l in filtered_features_nuts3[i] if l["properties"]["NUTS_ID"][:4] == k[:4]
            }
            for k in out_nuts2[i][j].keys()
        }
        for j in out_nuts2[i].keys()
    }
    for i in filtered_features_nuts2.keys()
}

region_ids_nuts3 = {
    country_code : [feature["properties"]["NUTS_ID"] for feature in ff]
    for country_code, ff in filtered_features_nuts3.items()
}

out_nuts3_name = {
    i : {
        j : {
            k : {
                l["properties"]["NAME_LATN"] : {
                    "type" : "Feature",
                    "properties" : l["properties"]
                }
                for l in filtered_features_nuts3[i] if l["properties"]["NUTS_ID"][:4] == k[:4]
            }
            for k in out_nuts2[i][j].keys()
        }
        for j in out_nuts2[i].keys()
    }
    for i in filtered_features_nuts2.keys()
}

region_names_nuts3 = {i : {
    j : {
        k : list(out_nuts3_name[i][j][k].keys())
        for k in out_nuts3_name[i][j].keys()
        }
    for j in out_nuts3_name[i].keys()
    }
 for i in out_nuts3_name.keys()
 }

## Save the formatted Geo data
json.dump(country_features_nuts3, 
          open("country_features_nuts3.json", 'w'))
json.dump(out_nuts3, 
          open("filtered_features_nuts3.json", 'w'))
json.dump(region_ids_nuts3, 
          open("region_ids_nuts3.json", 'w'))
json.dump(region_names_nuts3, 
          open("region_names_nuts3.json", 'w'))
```

```python
## ID to Name mapping

nuts1_name_map = {j["properties"]["NAME_LATN"] : j["properties"]["NUTS_ID"]
 for i in filtered_features_nuts1 for j in filtered_features_nuts1[i]}

nuts2_name_map = {j["properties"]["NAME_LATN"] : j["properties"]["NUTS_ID"]
 for i in filtered_features_nuts2 for j in filtered_features_nuts2[i]}

nuts3_name_map = {j["properties"]["NAME_LATN"] : j["properties"]["NUTS_ID"]
 for i in filtered_features_nuts3 for j in filtered_features_nuts3[i]}

nuts_name_map = {
    1 : nuts1_name_map,
    2 : nuts2_name_map,
    3 : nuts3_name_map
}

json.dump(nuts_name_map, 
          open("nuts_name_map.json", 'w'))
```
